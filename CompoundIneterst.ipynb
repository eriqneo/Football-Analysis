```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Function to calculate compound interest
def compound_interest(principal, rate, time):
    return principal * (1 + rate) ** time

# Parameters
principal_amount = 100000
daily_rate = 0.000219
days_in_year = 365
years = 1

# Create a DataFrame
data = {'Day': [], 'Balance': []}

for day in range(1, days_in_year + 1):
    balance = compound_interest(principal_amount, daily_rate, day)
    data['Day'].append(day)
    data['Balance'].append(balance)

df = pd.DataFrame(data)

# Display the DataFrame
df.tail()
```

```python
# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Day'], df['Balance'], label='Compound Interest')
plt.title('Compound Interest Over 1 Year')
plt.xlabel('Day')
plt.ylabel('Balance')
plt.legend()
plt.grid(True)
plt.show()
```
